```python
# Compound Interest Calculator

years = 2
rate = .03
principle = 3000

finalAmount = principle * (1 + rate) ** years

print(finalAmount)
```

```python
print(1,3)
# prints 1 and 3, "1 3\n" with a sep in the middle and end at the end so next line of code prints to next line
# sep = " "

print("1\n3")
# prints 1 above 3
# end = "\n" = start of next line###

```

```python
print("Principle:",principle)
print("Interest Rate:",rate)
print("Number of Years:",years)
#print function automatically adds an end "\n"

print("\nPrinciple:",principle, end="---")
print("Interest Rate:",rate, end="---")
print("Number of Years:",years)
#end can be replaced so new print calls print on the same line
```

```python
print("Interest Rate:", rate, sep=">>>", end="&&&")
print("Number of Years:", years)
#Can change sep and end to anything
```

```python
print("Interest Rate:", rate, "; Number of Years:", years)
#Can format in print
```

```python
print("Interest Rate: {} ; Number of Years: {}".format(rate, years))
# Or can format string itself before calling print
```
